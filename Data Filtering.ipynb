## Bivariate Analysis >> analysis of 2 columns

```python
# filter the rows having price greater than 30 and wieght greater than 3350

df[['Price','Weight']][(df['Price']>30)&(df['Weight']>3350)]
```

```python
# filter the rows having horsepower greater than 145 and having airbafs for driver and passenger both

# df.head(30) Driver & Passenger

df[['Horsepower','AirBags']][(df['Horsepower']>145)&(df['AirBags']=='Driver & Passenger')]
```

```python
# filter the rows having horsepower greater than 145 and having large or midsize type

df[['Horsepower','Type']][(df['Horsepower']>145)&((df['Type']=='Large')|(df['Type']=='Midsize'))]
```

```python
# filter the rows having price between 20 to 35 of large or midsize type.

df[['Price','Type']][((df['Price']>20)&(df['Price']<30))|((df['Type']=='Large')|(df['Type']=='Midsize'))]
```
