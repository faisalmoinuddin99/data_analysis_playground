```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Read the CSV file into a DataFrame

df = pd.read_csv("/home/kakashi/Data Analysis Playground/Population & Murder Rate - Sheet1.csv")
```

```python
df
```

```python
# Remove commas and convert the "Population" column to numeric

df["Population"] = df["Population"].replace({',':''}, regex=True).astype(float)
```

```python
df["Population"]
```

```python
df
```

```python
# Calculate the mean of the "Population" column/feature

mean_population = df["Population"].mean()
```

```python
print(f"Mean : {mean_population}")
```

```python
# Calculate the trimmed mean of the "Population" column with trim=0.1

trimmed_mean_population = stats.trim_mean(df["Population"],0.1)
```

```python
print(f"Trimmed Mean : {trimmed_mean_population}")
```

```python
median_population = df["Population"].median()
```

```python
print(f"Median : {median_population}")
```

```python
np.average(df["Murder rate"], weights = df["Population"])
```
