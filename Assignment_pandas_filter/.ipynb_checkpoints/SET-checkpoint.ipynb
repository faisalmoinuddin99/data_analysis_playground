# Pandas merge function

### inner merge

```python
pd.merge(emp,dept,how='inner',left_on='did',right_on='dept_id')
```

### Left Merge

```python
pd.merge(emp,dept, how='left', left_on='did', right_on='dept_id')
```

### Right Merge

```python
pd.merge(emp,dept, how='right',left_on='did', right_on='dept_id')
```

### Outer Merge

```python
pd.merge(emp,dept, how='outer',left_on='did', right_on='dept_id')
```
