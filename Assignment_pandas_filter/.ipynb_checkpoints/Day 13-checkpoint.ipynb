# Merge

```python
emp = pd.read_csv("/home/kakashi/Data Analysis Playground/data_analysis_playground/Assignment_pandas_filter/emp.csv")
```

```python
emp
```

```python
dept = pd.read_csv("/home/kakashi/Data Analysis Playground/data_analysis_playground/Assignment_pandas_filter/dept.csv")
```

```python
dept.head()
```

# Inner Merge - Intersection

```python
pd.merge(emp,dept,left_on='did',right_on='dept_id',how='inner') # will give common values
```

## Outer Merge - Union

```python
pd.merge(emp,dept,left_on='did',right_on='dept_id',how='outer')
```

## Right Merge

```python
pd.merge(emp,dept,left_on='did',right_on='dept_id',how='right')
```

## Left Merge

```python
pd.merge(emp,dept,left_on='did',right_on='dept_id',how='left')
```
