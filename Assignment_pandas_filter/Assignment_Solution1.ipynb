# group by >> category wise aggregation

```python
# what is the highest salary of each gender
df1.groupby(by='gender')['salary'].max()
```

```python
# what is the average salary of each gender
# what is the lowest salary of every gender
```

```python
df1.groupby(by='gender')['salary'].mean()
```

```python
df1.groupby(by='gender')['salary'].median()
```

```python
df1.groupby(by='gender')['salary'].min()
```
