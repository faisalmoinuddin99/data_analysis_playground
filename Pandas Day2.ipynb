```python
# add a column if price is less than 25 then 'cheap' if price is between 26 to 40 then 'moderate' and 'expensive' if
# price is greater than 40


df2.columns
```

```python
df2['Price_Category'] = df2['Price'].apply(
    lambda x: 'cheap' if x < 25 
    else ('moderate' if 26 <= x <= 40 else 'expensive'))
```

```python
d=[]

for i in df2["Price"]:
    if i < 25:
        d.append('cheap')
    elif i >= 25 and i <= 40:
        d.append('moderate')
    else:
        d.append('expensive')
```
